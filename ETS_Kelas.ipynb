# Missing Value dan Outlier

```python
data_mv = pd.read_csv('data_no_mv.csv')
data_mv.tail()
```

```python
from scipy import stats
data_mv_1=data_mv.drop(['Unnamed: 0','User_ID','Product_ID','Gender','Age','Occupation','City_Category','Stay_In_Current_City_Years','Marital_Status','Product_Category_1','Product_Category_2','Product_Category_3'],1)
z = np.abs(stats.zscore(data_mv_1._get_numeric_data()))
print(z)
data_mv_outlier=data_mv[(z<3).all(axis=1)]
print(data_mv_outlier.shape)
```

```python
data_mv_outlier.tail()
```

```python
data_mv_outlier_1=data_mv_outlier.drop(['Unnamed: 0','User_ID','Product_ID','Gender','Age','Occupation','City_Category','Stay_In_Current_City_Years','Marital_Status','Product_Category_1','Product_Category_2','Product_Category_3'],1)
data_mv_outlier_1.describe()
```

```python
data_mv_outlier.to_csv('data_mv_outlier.csv')
```

# Missing Value dan Transformasi

```python
data_mv = pd.read_csv('data_no_mv.csv')
data_mv.tail()
```

```python
from sklearn.preprocessing import normalize
data_mv_1=data_mv.drop(['Product_ID','Gender','Age','City_Category','Stay_In_Current_City_Years','Unnamed: 0','User_ID','Occupation','Marital_Status','Product_Category_1','Product_Category_2'],1)
data_mv_norm=normalize(data_mv_1,norm='l2') #default norm L2, other are L1 dan Max
data_mv_norm=pd.DataFrame(data_mv_norm)
data_mv_norm.tail()
```
