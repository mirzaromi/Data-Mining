```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
data_retail=pd.read_csv('online_retail.csv')
data_retail.head()
```

```python
data_retail.shape
```

```python
np.sum(data_retail.isnull())
```

```python
data_retail['UnitPrice'] = data_retail['UnitPrice'].astype('str')
data_retail = data_retail[~data_retail['UnitPrice'].str.contains('0')]
```

```python
np.sum(data_retail.isnull())
```

```python
data_retail_1 = data_retail.groupby('InvoiceNo')
transaksi = []
for i,j in data_retail_1:
    transaksi.append(list(j['Description'].map(str)))
```

```python
pd.DataFrame(transaksi)
```

```python
data_retail_2=pd.DataFrame(transaksi)
```

```python
data_retail_2.shape
```

```python
records=[]
for i in range (0, 22465):
    records.append([str(data_retail_2.values[i,j]) for j in range (0,923)])
```

```python
association_rules=apriori(records, min_support=0.0045,
                         min_confidence=0.2,
                         min_lift=3, min_length=2)
association_rules_result=list(association_rules)
```

```python
association_rules_result[0]
```

```python
for item in association_result:
#first index of the inner list
#contains base item and add item
    pair = item[0]
    items = [data_retail_1 for data_retail_1 in pair]
    print("rule: " + items [0] + " ->" + items [1])
    
    #second index of the inner list
    print("support: " + str(item[1]))
    
    #third index of the list located at 10th
    #of the third index of the inner list
    
    print("confidence: " + str(item[2][0][2]))
    print("lift: " + str(item[2][0][3]))
    print("=====================================")
```
